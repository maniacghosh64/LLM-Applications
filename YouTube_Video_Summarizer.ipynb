```python
from google.colab import userdata
GOOGLE_API_KEY = userdata.get("ANKAN_KEY")
```

Next, we extract YouTube transcript

```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_transcript(video_id):
  transcript = YouTubeTranscriptApi().fetch(video_id)
  print(transcript)
  text = " ".join([item.text for item in transcript])
  return text

# Correct usage
video_id = "bmvxRMYxlhE"
print(get_transcript(video_id))
```

```python
from google import genai

client = genai.Client(api_key="GOOGLE_API_KEY")
```

Finally we define the analyze/summary Function

```python
def analyze_video(transcript):
  prompt = f"""
  Analyze this youtube video transcript and provide:
  1. A short summary
  2. Detailed explanation
  3. Key points in a bullet list
  4. Important insights
  5. Questions and Answers

  Transcript: {transcript}
  """

  response = client.models.generate_content(
      model="gemini-2.5-flash-lite",
      contents=prompt
  )

  return response.text
```

```python
# Final call
video_id = "bmvxRMYxlhE"
transcript = get_transcript(video_id)
result = analyze_video(transcript)
print(result)
```

Chat Model

```python
def ask_question(transcript, question):
  prompt = f"""
  Answer the questions based on this video transcript:

  Transcript: {transcript}
  Question: {question}
  """

  response = client.models.generate_content(
      model='gemini-2.5-flash-lite',
      contents=prompt
  )

  return response.text

# Example
print(ask_question(transcript, "What is the main topic"))
```
