```python
# Read PDF
import PyPDF2

def read_resume(file_path):
  text = ""
  with open(file_path, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
      text += page.extract_text()
  return text

print(read_resume(file_name))
```

```python
# Generate with the help of Gemini using the 'client' connection so built
def analyze_resume(text):
  response = client.models.generate_content(
      model="gemini-2.5-flash-lite",
      contents=f"""
      You are an ATS resume analyzer.
      Analyze the resume below and return ATS score and feedback.
      Resume: {text}
      Also, return in JSON format with ats_scrore between 0-100,
      summary, strengths, weaknesses. Do not output any other text.
      """
  )
  return response.text
```
